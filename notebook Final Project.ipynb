### Analyse the Population 

Real Estate reasoning is simple but terribly effective: Supply and demand are at the heart of the real estate market!

It is therefore fundamental to try to anticipate the evolution of supply and demand to try to project the future dynamism of the real estate market:

If demand for housing increases, property prices increase if supply does not increase at the same rate;
If housing demand declines, property prices fall if housing supply does not fall at the same rate.
This is why it is necessary to know what is the growth of the population in the regions in which you want to invest.

```python
import requests
import pandas as pd

Data_link = 'https://www.ined.fr/en/everything_about_population/data/france/population-structure/regions_departments/'


html = requests.get(Data_link).content
df_list = pd.read_html(html)
df = df_list[0]
df.rename(columns={'Unnamed: 0': 'Region'}, inplace = True)
df.rename(columns={'Census 1990': 'Population (1990)'}, inplace = True)
df.rename(columns={'Census 1999': 'Population (1999)'}, inplace = True)
df.rename(columns={'January 1st 2008': 'Population (2008)'}, inplace = True)
df.rename(columns={'January 1st 2020 (p)': 'Population (2020)'}, inplace = True)

df
```

## Methodology <a name="methodology"></a>

The methods of expertise in matters of real estate and property valuation are

##### The income method:
It consists of taking either an existing income or a theoretical or potential income (rent versus market rental value), and then capitalizing it using a financial rate observed on the investment market. It is a method based on the future yield of a real estate asset. It is the extremely simplified version of the so-called cash flow approach (DCF).

##### The “soil and construction” method:
It consists in separately assessing the two components of the building, the land and the buildings. The value of the building (s) is assessed in consideration of a possible obsolescence deducting the new value. It is an extremely effective method as long as it is properly implemented.

##### The index method:
From an earlier value, it applies an index or coefficient of variation. This method is rarely used and formally discouraged in terms of market value. It is an old method and unsuitable for today's real estate markets.
